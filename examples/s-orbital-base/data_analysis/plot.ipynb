```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rs = np.linspace(0.5, 6, 100)
```

```python
rhf = np.load("../data/total_energies/rhf.npy")
uhf = np.load("../data/total_energies/uhf.npy")
ci = np.load("../data/total_energies/ci.npy")
```

```python
fig, ax = plt.subplots()
ax.plot(rs, rhf, label="RHF")
ax.plot(rs, uhf, label="UHF")
ax.plot(rs, ci, label="CI")
ax.set_xlabel("r")
ax.set_ylabel("E")
ax.axhline(0, color="black", linestyle="--")
ax.legend()
```
